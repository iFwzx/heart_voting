```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import ExtraTreesClassifier

ext = ExtraTreesClassifier(n_estimators=100, random_state=0)
dt = DecisionTreeClassifier(random_state=0)
knn = KNeighborsClassifier(n_neighbors=3)
gnb = GaussianNB()
log =  LogisticRegression()
svm = SVC()
sgd = SGDClassifier()
```

```python
def classifiers(clf1, clf2, clf3, clf4, clf5, clf6, clf7, X_train, y_train):
    clfs = [clf1, clf2, clf3, clf4, clf5, clf6, clf7]
    all_clfs_acc = []
    for clf in clfs:
        
        clf.fit(X_train, y_train)
        preds = clf.predict(X_train)
        acc = accuracy_score(y_train,preds)
        acc = acc.tolist()
        all_clfs_acc.append(acc)
        
    return all_clfs_acc
```

```python
classifiers(ext, dt, knn, gnb, log, svm, sgd, X_train, y_train)
```

```python
from sklearn.ensemble import VotingClassifier

vot_classifier = VotingClassifier(
    
    estimators=[('log1', log),
                ('svc1', svm),
                ('dt', dt),
               ('ext', ext),
               ('knn',knn),
                ('gnb',gnb),
                ('sgd', sgd)], 
    voting='hard')
```

```python
vot_classifier.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

def accuracy(model, data, labels):
    
    predictions = model.predict(data)
    acc = accuracy_score(labels, predictions)
    
    return acc
```

```python
accuracy(vot_classifier, X_train, y_train)
```
